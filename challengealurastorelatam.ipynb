#1. Análisis de facturación



```python
import matplotlib.pyplot as plt

def facturacion(tienda, numero_tienda):
  fac = round(tienda['Precio'].sum(), 2)
  print(f'{numero_tienda} Facturo ${fac}')
  return fac
```

```python
fac1 = facturacion(tienda, 'tienda 1')
fac2 = facturacion(tienda2, 'tienda 2')
fac3 = facturacion(tienda3, 'tienda 3')
fac4 = facturacion(tienda4, 'tienda 4')

nombreTiendas = ['tienda 1', 'tienda 2', 'tienda 3', 'tienda 4']
facturacion = [fac1, fac2, fac3, fac4]

plt.bar(nombreTiendas, facturacion)
plt.title("Facturacion por Tienda")
plt.ylabel("Monto de Ventas")
plt.tight_layout()
plt.show()
```

```python
def mostrar_total(pct, all_vals):
    valor_absoluto = int(round(pct/100. * sum(all_vals)))
    porcentaje = f'{pct:.1f}%'  # Formato con 1 decimal
    return f'{porcentaje}\n({valor_absoluto})'  # Mostrar % y valor absoluto

categoria_1 = tienda['Categoría del Producto'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(categoria_1,
        labels=categoria_1.index,
        autopct=lambda x: mostrar_total(x, categoria_1),
        startangle=90,
        wedgeprops={'linewidth': 1, 'edgecolor': 'white'})
plt.title('Ventas por Categoría - Tienda 1', pad=20)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
media_1 = round(float(tienda['Calificación'].sum() / len(tienda)), 2)
media_2 = round(float(tienda2['Calificación'].sum() / len(tienda2)), 2)
media_3 = round(float(tienda3['Calificación'].sum() / len(tienda3)), 2)
media_4 = round(float(tienda4['Calificación'].sum() / len(tienda4)), 2)

print(f'tienda 1 tiene una media de {media_1} estrellas')
print(f'tienda 2 tiene una media de {media_2} estrellas')
print(f'tienda 3 tiene una media de {media_3} estrellas')
print(f'tienda 4 tiene una media de {media_4} estrellas')

plt.barh(['Tienda 4', 'Tienda 3', 'Tienda 2', 'Tienda 1'], [media_4, media_3, media_2, media_1])
plt.xlabel('Calificación')
plt.ylabel('tienda')
plt.title('Calificación media por tienda')
for i, valor in enumerate([media_4, media_3, media_3, media_1]):
  plt.text(valor, i, f'{valor}', va='center', ha='left')

plt.show()
```

```python

# Asumiendo que los datos ya están importados como:
# tienda, tienda2, tienda3, tienda4

# Configuración
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
datos_tiendas = [tienda, tienda2, tienda3, tienda4]
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Colores distintivos

# Procesar los datos
resultados = []
for nombre, df in zip(nombres_tiendas, datos_tiendas):
    # Calcular métricas
    promedio = df['Calificación'].mean()
    conteo = df['Calificación'].value_counts().sort_index()
    porcentajes = (conteo / len(df)) * 100

    resultados.append({
        'nombre': nombre,
        'promedio': promedio,
        'conteo': conteo,
        'porcentajes': porcentajes,
        'total_productos': len(df)
    })

# Crear figura con dos subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10),
                         gridspec_kw={'height_ratios': [1, 3]})

# Gráfico 1: Promedios de calificación
ax1.bar([r['nombre'] for r in resultados],
        [r['promedio'] for r in resultados],
        color=colores)
ax1.set_title('Comparación de Calificaciones Promedio', fontsize=14, pad=20)
ax1.set_ylabel('Promedio (1-5)')
ax1.set_ylim(0, 5.5)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir valores en las barras
for i, r in enumerate(resultados):
    ax1.text(i, r['promedio'] + 0.1, f"{r['promedio']:.2f}",
            ha='center', va='bottom', fontsize=10, fontweight='bold')

# Gráfico 2: Distribución detallada
ancho_barra = 0.2
calificaciones = np.arange(1, 6)  # Asumimos calificaciones de 1 a 5

for i, r in enumerate(resultados):
    posiciones = calificaciones + i * ancho_barra
    ax2.bar(posiciones, r['conteo'], width=ancho_barra,
            color=colores[i],
            label=f"{r['nombre']} (n={r['total_productos']})")

    # Añadir porcentajes
    for j, cal in enumerate(calificaciones):
        if cal in r['conteo']:
            ax2.text(posiciones[j], r['conteo'][cal] + 5,
                    f"{r['porcentajes'][cal]:.1f}%\n({r['conteo'][cal]})",
                    ha='center', fontsize=8)

ax2.set_title('Distribución de Calificaciones por Tienda', fontsize=14, pad=20)
ax2.set_xlabel('Calificación (1-5)')
ax2.set_ylabel('Cantidad de Productos')
ax2.set_xticks(calificaciones + ancho_barra * (len(resultados)-1)/2)
ax2.set_xticklabels(calificaciones)
ax2.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
ax2.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar layout y mostrar
plt.tight_layout()
plt.show()

# Mostrar resumen en consola
print("\nResumen de Calificaciones:")
print("{:<10} {:<10} {:<10}".format('Tienda', 'Promedio', 'Productos'))
for r in resultados:
    print("{:<10} {:<10.2f} {:<10}".format(r['nombre'], r['promedio'], r['total_productos']))
```

# 4. Productos más y menos vendidos

```python

# 1. Preparación de datos combinados
# Agregar columna de tienda a cada DataFrame
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Combinar todos los datos
datos_completos = pd.concat([tienda, tienda2, tienda3, tienda4])

# 2. Análisis de productos más y menos vendidos
# Contar ventas por producto
ventas_por_producto = datos_completos['Producto'].value_counts().reset_index()
ventas_por_producto.columns = ['Producto', 'Ventas']

# Productos más vendidos (top 10)
top_productos = ventas_por_producto.head(10)

# Productos menos vendidos (bottom 10)
bottom_productos = ventas_por_producto.tail(10)

# 3. Visualización
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))

# Gráfico de productos más vendidos
ax1.barh(top_productos['Producto'], top_productos['Ventas'], color='green')
ax1.set_title('Top 10 Productos Más Vendidos', pad=20, fontsize=14)
ax1.set_xlabel('Cantidad de Ventas')
ax1.invert_yaxis()  # Para mostrar el más vendido arriba

# Añadir etiquetas de cantidad
for i, v in enumerate(top_productos['Ventas']):
    ax1.text(v + 1, i, str(v), color='green', fontweight='bold')

# Gráfico de productos menos vendidos
ax2.barh(bottom_productos['Producto'], bottom_productos['Ventas'], color='red')
ax2.set_title('Top 10 Productos Menos Vendidos', pad=20, fontsize=14)
ax2.set_xlabel('Cantidad de Ventas')
ax2.invert_yaxis()  # Para mostrar el menos vendido arriba

# Añadir etiquetas de cantidad
for i, v in enumerate(bottom_productos['Ventas']):
    ax2.text(v + 0.5, i, str(v), color='red', fontweight='bold')

plt.tight_layout()
plt.show()

# 4. Análisis por tienda (opcional)
fig2, axes = plt.subplots(2, 2, figsize=(15, 12))
fig2.suptitle('Productos Más Vendidos por Tienda', fontsize=16)

tiendas = {'Tienda 1': tienda, 'Tienda 2': tienda2,
           'Tienda 3': tienda3, 'Tienda 4': tienda4}

for ax, (nombre, df) in zip(axes.flatten(), tiendas.items()):
    top_tienda = df['Producto'].value_counts().head(5)
    top_tienda.plot(kind='barh', ax=ax, color='skyblue')
    ax.set_title(nombre)
    ax.set_xlabel('Ventas')
    ax.invert_yaxis()

    for i, v in enumerate(top_tienda):
        ax.text(v + 0.5, i, str(v), color='blue', fontweight='bold')

plt.tight_layout()
plt.show()

# 5. Mostrar resultados en consola
print("\nProducto más vendido en general:")
print(ventas_por_producto.iloc[0])

print("\nProducto menos vendido en general:")
print(ventas_por_producto.iloc[-1])

print("\nResumen por tienda:")
for nombre, df in tiendas.items():
    top = df['Producto'].value_counts().head(1)
    print(f"\n{nombre}:")
    print(f"Producto más vendido: {top.index[0]} ({top.values[0]} ventas)")
```

# 5. Envío promedio por tienda

```python

costos_promedio = {
    'Tienda 1': tienda['Costo de envío'].mean(),
    'Tienda 2': tienda2['Costo de envío'].mean(),
    'Tienda 3': tienda3['Costo de envío'].mean(),
    'Tienda 4': tienda4['Costo de envío'].mean()
}

# 2. Crear DataFrame con los resultados
df_costos = pd.DataFrame.from_dict(costos_promedio, orient='index', columns=['Costo Promedio'])
df_costos = df_costos.sort_values('Costo Promedio', ascending=False)
# Mostrar tabla con los datos
print("\nCosto de envío promedio por tienda:")
print(df_costos.to_markdown(tablefmt="grid", floatfmt=",.0f"))
```

```python

# 3. Visualización
plt.figure(figsize=(10, 6))
bars = plt.bar(df_costos.index, df_costos['Costo Promedio'],
               color=['#4C72B0', '#55A868', '#C44E52', '#8172B2'])

# Personalizar gráfico
plt.title('Costo Promedio de Envío por Tienda', fontsize=16, pad=20)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Costo Promedio (COP)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir etiquetas con los valores
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'${height:,.0f}',
             ha='center', va='bottom', fontsize=11)


plt.tight_layout()
plt.show()
```
